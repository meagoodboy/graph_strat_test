```python
def process_data(data, clinical,clin_param):
    
    data = data[~data.index.duplicated(keep='first')]
    pat = list(set(clinical.index).intersection(set(data.index)))
    # print(pat)
    data = data.loc[pat,:]
    clinical = clinical.loc[pat,:]
    
    # finding uniqe clinical params
    y = list(set(clinical[clin_param]))
    
    y = [str(i) for i in y]
    y.sort()
    print(y)
    
    # finding common patients
    clin_pat = []
    for i in clinical.index:
        if clinical.loc[i,clin_param] in y:
            clin_pat.append(i)
    
    data = data.loc[clin_pat,:]
    clinical = clinical.loc[clin_pat,:]
    
    # Creating a label
    feature = []
    for i in clinical[clin_param]:
        for j in range(len(y)):
            if i == y[j]:
                feature.append(j+1)
                
    return data , feature
```

```python
data, features = process_data(data, clinical, "ajcc_pathologic_stage")
```

```python
# umap
umap_reducer = umap.UMAP()
embedding_umap = umap_reducer.fit_transform(data)
umap_emb = pd.DataFrame(embedding_umap,columns=['x','y'])
umap_emb['label'] = features
umap_plot = sns.scatterplot(data=umap_emb, x='x', y='y', hue='label',s=70, alpha=0.8, palette="coolwarm")
umap_plot.set(title="UMAP")
fig = umap_plot.get_figure()
fig.savefig(save_name+ "_umap.png") 
```

```python
# pca
pca_reducer = PCA(n_components=5)
embedding_pca = pca_reducer.fit_transform(data)
pca_emb = pd.DataFrame(embedding_pca,columns=['x','y','z','a','b'])
pca_emb['label'] = features
pca_plot = sns.scatterplot(data=pca_emb, x='x', y='y', hue='label',s=70, alpha=0.8, palette="coolwarm")
pca_plot.set(title='PCA', xlabel = str(pca_reducer.explained_variance_ratio_[0])[:5] , ylabel = str(pca_reducer.explained_variance_ratio_[1])[:5])
fig = pca_plot.get_figure()
fig.savefig(save_name+ "_pca.png") 
```

```python
# tsne
tsne_reducer = TSNE(2, learning_rate='auto', init='random')
embedding_tsne = tsne_reducer.fit_transform(data)
tsne_emb = pd.DataFrame(embedding_tsne,columns=['x','y'])
tsne_emb['label'] = features
tsne_plot = sns.scatterplot(data=tsne_emb, x='x', y='y', hue='label',s=70, alpha=0.8, palette="coolwarm")
tsne_plot.set(title='TSNE', xlabel=str(tsne_reducer.kl_divergence_)[:5])
fig = tsne_plot.get_figure()
fig.savefig(save_name+ "_tsne.png") 
```
