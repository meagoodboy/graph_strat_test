```python
import pandas as pd
```

```python
enc_file = "raw/kirp_sm_miss251.csv"
output_file = enc_file[:-4]
```

```python
enc = pd.read_csv(enc_file, index_col=0)
```

```python
mad_val = [500]

x = enc.mad(axis=0, skipna=None, level=None)
x = pd.DataFrame(x)
x = x.sort_values(0,ascending=False)

for i in mad_val:
    mad_genes = x.head(i).index
    print("Filtered by mad values")
    df = enc[mad_genes]
    # normalized_df=(df-df.min())/(df.max()-df.min())
    df.to_csv(output_file+"_"+str(i)+"_mad.csv" )
```
